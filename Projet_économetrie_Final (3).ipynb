Les économies du monde seraient « held back by uncertainty » (1)  : c’est la position que défend l’IMF pour expliquer la lenteur de la stabilisation des économies et des marchés mondiaux suite à la crise des sub-primes. Ainsi, l’incertitude concernant les politiques fiscales, régulatoires et économiques est devenue depuis ces crises financières du début du 21eme siècle, qui ont bouleversées les économies mondiales, une préoccupation majeure.  (2)

L’Europe a en particulier été un théâtre de calamités financières. La crise de 2008 a été rapidement suivie par la crise de la zone euro en 2011 qui est née en Grèce mais s’est rapidement répandue dans l’ensemble de la zone. La dette publique a atteint des niveaux colossaux et la volatilité sur les marchés européens a connu une hausse importante. La crise a suscité des débats et des négociations parmi les dirigeants européens sur la stabilité de l'euro et la nécessité d'une assistance financière pour les pays en difficulté. Tous les yeux étaient tournés vers la Banque centrale européenne (BCE) et le Fonds monétaire international (FMI), ainsi que d’autres régulateurs pour l’établissement d’un plan de secours et une perspective de fin de la crise. L’incertitude et la crainte ont joué un rôle clé dans la propagation de la crise : les investisseurs sont devenus méfiants à l'égard de la détention de dette souveraine de pays jugés à risque élevé.

Dans ce contexte, il nous a paru intéressant d’étudier la relation entre l'incertitude portant sur les politiques économiques et les dynamiques du marché obligataire en Europe. Le stress souverain est important à quantifier, car les tensions sur les marchés obligataires souverains, en raison de leur taille considérable, peuvent facilement se propager à d'autres segments du marché et ainsi créer des risques systémiques pour l'ensemble du système financier.

Notre étude vise donc à comprendre de quelle manière l’incertitude portant sur les politiques économiques impacte le niveau de stress systémique en Europe.

Pour ce faire, nous avons sélectionné les variables suivantes (3) :
-	**Variable ENI** (European News Index) : Une variable comportant les données relatives à un indice composite qui mesure l’incertitude portant sur les politiques économiques en Europe entre Septembre 2000 et Septembre 2023
Les données sont tirées des principales médias européens (Le Monde, Le Figaro, Handelsblatt, La Stampa, El Mundo, The Financial Times, etc.). Les auteurs de l’indice comptent le nombre d’apparitions du mot « uncertain » ou « uncertainty », « economic » ou « economy », et autres termes associés.

 L’incertitude porte sur les questions suivantes  :

•	Qui va prendre les décisions économiques ?

•	En quoi vont consister ces politiques et quand vont-elles être mises en place ?

•	Quels vont être les effets ?

- **Variable STRESS INDEX** : Une variable comportant les données relatives à un indice composite qui mesure le stress systémique souverain en Europe entre Septembre 2000 et Septembre 2023.
Pour donner un bref aperçu de sa construction, on peut dire que le stress sur les marchés souverains est le plus souvent mesuré par l'écart de rendement entre les obligations d'un pays et une valeur refuge sûre ("spread de crédit"). Cependant, il peut également se manifester par une volatilité des prix plus élevée et une liquidité plus faible dans le trading d'obligations ; incorporer ces deux symptômes de stress, ainsi qu'un spread de crédit, dans une mesure globale (composite) du stress sur les marchés souverains a donc du sens pour commencer. Les auteurs ont donc incorporé dans l’indice trois dimensions :

(i)	les écarts de risque captent la tarification du risque de crédit et de
liquidité des obligations sous-jacentes

(ii)	les volatilités des rendements qui mesurent le degré de fragilité et d'incertitude prévalant sur le marché

(iii)	les écarts entre les cours acheteur et vendeur qui reflètent les coûts de transaction et les conditions de liquidité du marché de manière plus générale. (4)

Nous nous attendons à ce qu’il y ait une corrélation positive entre ces deux variables, et un effet feedback entre les variable dans le modèle multivarié (compte tenu de ce qui a été dit ci-dessus).

Il semble logique de considérer qu’une crise financière impacte le niveau d’incertitude économique  (la situation sur le marché répand le doute quant à la réaction des régulateurs) tout comme on pourrait aisément  penser l’inverse (le doute sur les politiques à venir modifie les comportements des acteurs du marché, comme indiquent de la finance comportementale ). On s'attend tout de même à ce que la variable causale, exogène, soit celle portant sur l'incertitude. Nous voudrions donc vérifier s'il y a bien une causalité et  déterminer si oui ou non l'incertitude est réellement un frein pour les marchés.


Nous allons donc poursuivre notre étude pour confirmer, ou infirmer, nos hypothèses.

# **Question 3 : Test de racine unitaire et méthode de stationnarisation approporiée**


L'analyse des autocorrélogramme suggère que nos séries ne sont pas stationnaires. En l'état, nous ne pouvons les modéliser. Nous effectuons des test de stationnarité afin de rationaliser notre anlayse, et choisir une méthode de stationarisation adaptée.

### a) Indice d'incertitude des politiques économiques










### Test de Dickey-Fuller augmenté

#### Test ADF avec constante et tendance déterministe (M3)


On applique le test ADF avec une tendance déterministe et une constante. Le nombre de retards maximum est fixé à 10 par convention et le nombre de retards optimal est déterminé par la minimisation du critère AIC.

```r
#Importation des packages
install.packages("urca")
library(urca)

#M3
adf_M3 <- ur.df(uncertainty_data$logENI, type = "trend", lags = 10 , selectlags = "AIC")
summary(adf_M3)
```

```r
#Calcul du seuil du TM3
#On utilise les tables de la loi normale du fait du nombre d'observations
seuil_rejet <- qt(1 - 0.05/2, 277)
print(seuil_rejet)
```

**TRU :**
La statistique du test ADF est égale à  $t_{ADF}=-4.012$. Elle est inférieure à la valeur critique au seuil de 5\% $t_{ADF}=-4,12<-3.42$, mais aussi au seuil de 1% (-3,98)et 10% (-3,13).  
**On rejette donc l'hypothèse nulle de racine unitaire, ie de non stationnarisé.**

**TM3 :**
La statistique de test est égale à $t_{\hat{b}} =3,18$. Comme on a rejetté H0 pour le TRU, on lit le seuil de rejet qui figure sur les tables de la loi Normale pour un degré de liberté de 266 et un risque de première espèce de 5%. On peut aussi le faire grâce à la fonction qt(). Il est égal à 1,97 pour un risque de première espèce de 5%. On a donc  $\vert t_{\hat{b}} \vert =3,18>1,97$ : on rejette  l'hypothèse nulle que la tendance déterministe n'est pas significative.


**Nous concluons à un modèle  I(0)+C+T.**
**Il s'agit d'un procéssus TS. La non-stationarité est détérministe.**

Le résultat du test ADF indique que un choc sur l'indice d'incertitude est  transitoire. Ceci semble cohérent. On pourrait penser par exemple qu'un choc sur l'incertitude ne peut durer trop longtemps, car au niveau de l'UE, il y a un organe décisionnel très puissant et une réglementation très forte qui contraint les gouvernments et les acteurs concernés à prendre des décisions et de les implémenter assez rapidement. Les chocs sur cette série sont dus à des événements spécifiques dont les effets sont forcément atténués par le temps.

### Test de KPSS

Les tests minimisent en priorité le risque de 1re espèce
i.e. dans les tests précédents le rejet à tort de l’hypothèse
nulle de RU. On souhaite à présent limiter en priorité le risque
de rejeter à tort l’absence de RU. Pour cela, on inverse les
hypothèses dans le test KPSS et contre-testons nos résultats.

```r
#Test de stationnarité autour d'une tendance deterministe

kpssM3 <- ur.kpss(uncertainty_data$logENI, type = 'tau', lags = 'long')
summary(kpssM3)
```

Nous avons des données mensuelles donc nous choisissons de rajouter plus de lag pour plus de clarté dans la conclusion, notamment au niveau de la cohérence entre les tests.

La statistique du test est égale à $LM_{KPSS} = 0.097$. Le seuil de rejet pour un risque de première espèce est égal à 0.146 au seuil de 5%. On a donc $LM_{KPSS} = 0,097<0,146$.

On ne peut pas rejeter l'hypothèse nulle d'abscence de RU au seuil de 5%, ce qui est conforme à notre résultat précédent.

### Stationnarisation

Pour stationnariser notre série et au vu des résultats précédents, nous choisissons de regresser les observations sur un index pour éliminer la tendance.

```r
#Stationarisation
index <- 1:277
modelENI <- lm(uncertainty_data$logENI~index)
summary(modelENI)
z<-modelENI$residuals
```

Avec cette méthode, nous capturons les résidus, qui représentent ce qu'il reste de notre série temporelle lorsqu'on enlève la tendance déterministe, dans la variable z.

Nous recréons un dataframe à partir de la série stationnarisée et vérifions que tout reste cohérent.

```r
# Création d'un dataframe avec les dates et les résidus
df_stationnaire <- data.frame(date = uncertainty_data$date, ENI_stationnaire = z)

# Affichage des premières lignes du dataframe
head(df_stationnaire)
```

```r
#Représentation graphique de la série stationnarisée

p_s<-ggplot(data=df_stationnaire,aes(x=date,y=ENI_stationnaire))+geom_line(color="red")+ggtitle("Résidus stationnaires de la régression",subtitle="Septembre 2000-Septembre 2023")
print(p_s)
```

Pour s'assurer que notre stationarisation est suffisante, nous appliquons un test ADF à la série stationnaisée pour s'assurer qu'elle soit bien stationnaire.



```r
adf_M3bis <- ur.df(z, type = "trend", lags = 10, selectlags = "AIC")
summary(adf_M3bis)
```

```r
adf_M2bis <- ur.df(z, type = "drift", lags = 10, selectlags = "AIC")
summary(adf_M2bis)
```

```r
adf_M1bis <- ur.df(z,type="none", lags=10, selectlags = "AIC")
summary(adf_M1bis)
```

Les conclusions aux trois test sont les suivantes :

**M3**

*TRU* : rejet H0 (t = -4.012 < c =-3,42)

*TM3* : non rejet H0 ( t = 0,047< u = 1,97)

**M2**

*TRU* : rejet H0 (t = -4.019 < c = -2.87)

*TM2* : non rejet H0 ( t = -0,048 < u = 1,97)

**M1**

*TRU* : rejet H0 (t = -4.02 < c = -1.95)

Le procéssus est stationnaire et sans constante.

On refait le processus de test RU pour l'autre série.

On peut donc conclure qu'il s'agit d'un procéssus DS sans dérive. Autrement dit, la série temporelle n'est pas stationnaire à cause de variations aléatoires ou stochastiques dans le temps, pas à cause d'une tendance ou dérive. L'absence de dérive suggère que la série ne présente pas de tendance linéaire constante. Autrement dit, par de hausse ou baisse systématiques des observations dans le temps.

Un procéssus DS est caractérisé par sa "mémoire infinie des chocs". Si le choc entraîne des changements structurels dans les marchés obligataires souverains, tels que des modifications de la réglementation, de la dynamique du marché ou du comportement des investisseurs, les effets du choc peuvent en effet persister à long terme. D'autre part, nous avons pris des données de 2000-2023 qui prennent en compte la crise de 2011 qui s'est en effet durablement imprégné de la structure et du fonctionnement des marchés souverains.


**La non-stationarité est stochastique. Il faut donc différencier la série pour la stationnariser.**

Nous vérifions notre conclusion à l'aide d'un test KPSS.

### Test de KPSS

```r
kpssM1<- ur.kpss(stressindex_data$index, type = 'mu', lags = 'short')
summary(kpssM1)
```

La statistique du test est égale à $LM_{KPSS} = 0.59$. Le seuil de rejet pour un risque de première espèce est égal à 0.463 au seuil de 5%. On a donc $LM_{KPSS} = 0.59>0,463$.

On peut rejeter l'hypothèse nulle de stationarité au seuil 5%. Ces résultats sont cohérents avec ceux de nos test ADF.

### Stationnarisation

Nous commencons la stationnarisation par une simple différenciation, qui consiste à soustraire chaque observation par la précédente. Avec cette méthode, on supprime les variations stochastiques qui rendent la série non-stationnaire, c'est à dire les composants dépendant du temps, tout en évitant d'introduire des dépendances artificielles dans la série.


```r
# Création d'une variable date
dates <- stressindex_data$date

# Différenciation de la série et création d'une nouvelle série temporelle
D_stressindex_data <- diff(stressindex_data$index)

# Créer un vecteur de dates pour la série différenciée (sans la première date)
dates_diff <- dates[-1]

# Tracer la nouvelle série temporelle différenciée avec ggplot2
p_v_diff <- ggplot(data = data.frame(date = dates_diff, index_diff = D_stressindex_data),
                   aes(x = date, y = index_diff)) +
  geom_line(color = "blue") +
  ggtitle("Différence première de l'indice de volatilité") +
  theme(plot.title = element_text(face = "bold"))

print(p_v_diff)
```

Ce graphique suggère que la série différenciée une fois est bien stationnaire. Vérifions avec le test ADF et le test KPSS.

```r
adf_M1bis <- ur.df(y = D_stressindex_data,type="none", lag=10, selectlags = 'AIC')
summary(adf_M1bis)
```

On a bien t*obs* = -6,1239 < -1,95. Donc on rejette l'hypothèse H0 avec un risque de première espèce de 5%. La série est bien différenciée et stationnaire.


```r
kpssM1<- ur.kpss(D_stressindex_data, type = 'mu', lags = 'long')
summary(kpssM1)
```

La statistique du test est égale à $LM_{KPSS} = 0.0654$. Le seuil de rejet pour un risque de première espèce est égal à 0.463 au seuil de 5%. On a donc $LM_{KPSS} = 0.0654<0,463$.

On ne rejette pas l'hypothèse nulle de stationarité au seuil 5%.

### b) Méthode alternative pour confirmer nos hypothèses : les critères AIC et BIC




Pour confirmer notre lecture des autocorrélogrammes simples et partiels, nous utilisons une méthode alternative en créant une boucle qui calcule les critères d'information AIC et BIC pour différents retards p et q. Nous devons retenir les retards p et q pour lesquels ces critères sont minimisés.

```r

install.packages("forecast")
library(forecast)
```

```r
#Choix de l'ordre avec les critères d'informations
mat_aic <- matrix(0, nrow = 5, ncol = 5)
mat_bic <- matrix(0, nrow = 5, ncol = 5)


for (i in 0:4) {
  for (j in 0:4) {
    ARMA_ij <- Arima(y = df_diff$index_diff, order = c(i, 0, j), method = "ML")
    mat_aic[i + 1, j + 1] <- ARMA_ij$aic
    mat_bic[i + 1, j + 1] <- ARMA_ij$bic
  }
}

rownames(mat_aic) <- c('p=0', 'p=1', 'p=2', 'p=3', 'p=4')
colnames(mat_aic) <- c('q=0', 'q=1', 'q=2', 'q=3', 'q=4')
rownames(mat_bic) <- c('p=0', 'p=1', 'p=2', 'p=3', 'p=4')
colnames(mat_bic) <- c('q=0', 'q=1', 'q=2', 'q=3', 'q=4')
print(mat_aic)
```

Les processus qui minimisent le critère AIC sont (dans l'ordre) les ARMA(3,3), ARMA (4,3), ARMA (1, 1) et AR(4).  Nous ne retrouvons pas le processus AR (1) identifié en étudiant les autocorrélogrammes partiels.

```r
print(mat_bic)
```

Le processus qui minimise le critère BIC est un AR(1), le processus que nous avions identifié en étudiant les autocorrélogrammes.

Le fait que les deux critères ne donnent pas les mêmes résultats n'est pas normal. Nous ne parvenons pas à trouver l'erreur. On procède donc avec les deux modèles sélectionnés pour choisir le meilleur.

# **Question 8 : Analyse impulsion-réponse des chocs sur les différentes variables**

Nous allons procéder à l'analyse impulsion-réponse qui permet de comprendre l'impact d'un choc sur les variables.

Pour cela nous utilisons la décomposition de Cholesky qui permet d’obtenir des chocs
orthogonaux. La décomposition exige d’ordonner les variables du système selon un degré
d’exogénéité décroissant. Les résultats des tests de causalité à la Granger ne nous font  
rejeter l’hypothèse d'absence de causalité pour les deux variables, donc pour savoir laquelle est
la plus exogène nous comparons les p-value des tests :

- test ENI sur stress index : p-value = 0.1666

- test stress index sur ENI : p-value = 0.69

Comme 0.69>0.16, on peut dire que l'on est plus ferme dans le non- rejet de l'hypothèse de non-causalité sur ENI. On la place donc au-dessus du stress index. C'est la variable la plus exogène, la moins causée par l'autre. Ceci semble coherent : c'est l'incertitude relative aux politiques économiques qui se répercute sur le marché plutôt que l'inverse.

```r
var.1lag <-VAR(y = df_sans_dates[,c("ENI_stationnaire","index_diff")],type="none",lag.max = 1)
```

```r
dec_var <-fevd(var.1lag,n.ahead=20)

plot(dec_var)
```

Quelque soit l'horizon de prévision considéré, la variance de l'erreur de prévision de la premiere variable du VAR (ENI) est expliquée à 100% par les chocs sur elle-même, tandis que la deuxième (index) est expliquée principalement par les chocs sur elle-même et marginalement par les chocs sur la première.

**Méthode VAR**



```r
irf<-irf(var.1lag,n.ahead = 12,ortho = TRUE,ci=0.95)
plot(irf)
```

Le premier graphique montre l'impact d'un choc sur ENI et le deuxième un choc sur le stress index.

L'analyse des graphiques nous permet de tirer les conclusions suivantes :

- **impact d'un choc sur ENI sur**

➡ lui-même : l'impact est immédiat, négatif. Il se prolonge dans le temps avant de s'estomper.

➡ le stress index : l'impact n'est pas immediat, est de petite ampleur et négatif, et s'éstompe rapidement.

- **impact d'un choc sur le stress index sur**

➡ lui-même : l'impact est immédiat, négatif. Il s'estompe rapidement.

➡ ENI : l'impact n'est pas immédiat, il est positif. Il s'estompe lentement.

Conformément aux résultats des tests de causalité, les fonctions d'impulsion réponses ne sont pas significatifs.

On note également dans les 4 graphiques une convergence vers 0 de la fonction de réponse ce qui permet de conforter la conslusion sur la stationarité du VAR.


**Méthode des projections locales**


```r
library(lpirfs)

irf_LP <-lp_lin(endog_data = df_sans_dates[, c(2, 1)],
                    lags_endog_lin = NaN,
                    lags_criterion = 'BIC',
                    max_lags = 12,
                    trend = 0,
                    shock_type = 0,
                    confint=1.96,
                    hor=12,
                    adjust_se=TRUE)

plot(irf_LP)
```

Les conclusions sont les suivantes :

- **impact d'un choc sur ENI sur**

➡ lui-même : conclusion identique à celle tirée de la méthode VAR

➡ le stress index : conclusion identique à celle tirée de la méthode VAR

- **impact d'un choc sur le stress index sur**

➡ lui-même : conclusion identique à celle tirée de la méthode VAR

➡ ENI : l'impact n'est pas immédiat, positif. Cependant, l'impact positif dure jusqu'au deuxième mois post-choc où il devient négatif. Cet effet négatif n'est pas observable avec la méthode VAR.

De nouveau, les réponses aux chocs ne sont pas significatives.

Il faut aussi dire que les durées des chocs sont plus importants avec cette méthode. Ceci peut s'expliquer par le fait que les fonctions impulsion réponse sont estimées en déterminant de façon endogène le nombre de retards pour chaque horizon.
